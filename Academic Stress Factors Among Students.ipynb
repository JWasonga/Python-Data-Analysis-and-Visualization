###**Histogram Plots - To inspect Distributions Academic performance, Stress Levels, sleep quality rate etc.**

```python
cols = ["Kindly Rate your Sleep Quality", "How would you rate your stress levels?", "How would you rate you academic performance"]

fig, axes = plt.subplots(1, 3, figsize=(12, 4))

for ax, col in zip(axes.ravel(), cols):
    sns.histplot(df[col], ax=ax, kde=False)
    ax.set_title(col) # Changed cols to col to display the title of each subplot correctly
plt.suptitle("Distribution of Sleep Quality, Stress Levels and Academic Performance")
plt.tight_layout()
plt.show()
```

###**Pair Plots - To inspect relationships**

```python

sns.set_style('white')
g = sns.pairplot(df, diag_kind="kde", height=3)
plt.tight_layout()
plt.subplots_adjust(wspace=0.2, )
```
